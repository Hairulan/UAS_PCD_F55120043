```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
orang=cv2.imread('NarutoUzumaki.png')
orang=cv2.cvtColor(orang,cv2.COLOR_BGR2RGB)
hsv_orang=cv2.cvtColor(orang, cv2.COLOR_RGB2HSV)
```

```python
light_orange=(1,190,200)
dark_orange =(18, 255, 255)
```

```python
mask=cv2.inRange(hsv_orang,light_orange,dark_orange)
result=cv2.bitwise_and(orang,orang, mask=mask)
```

```python
plt.subplot(1,4,1)
plt.imshow(orang)
plt.subplot(1,4,2)
plt.imshow(hsv_orang)
plt.subplot(1,4,3)
plt.imshow(mask)
plt.subplot(1,4,4)
plt.imshow(result)
plt.show()
```
